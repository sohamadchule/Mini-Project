```python
import kagglehub

path = kagglehub.dataset_download("shudhanshusingh/az-medicine-dataset-of-india")

print("Path to dataset files:", path)
```

```python
import os

# List all files in the downloaded dataset directory
files = os.listdir(path)
print("Files in the dataset folder:", files)
```

```python
import pandas as pd

# Construct the full path to the CSV file
csv_file = 'A_Z_medicines_dataset_of_India.csv'
csv_path = os.path.join(path, csv_file)

# Load the CSV into a DataFrame
df = pd.read_csv(csv_path)
```

```python
df
```

```python
# Clean the 'name' column to remove NaN values, strip spaces, and convert to lowercase
medicine_names = df['name'].dropna().str.lower().str.strip().tolist()

# Display the first few cleaned names
medicine_names[:10]
```

```python
# Drop columns that are not needed for the suggestion system
df_cleaned = df.drop(columns=['id', 'price(₹)', 'Is_discontinued', 'manufacturer_name', 'type', 'pack_size_label', 'short_composition1', 'short_composition2'])

# Display the cleaned DataFrame to confirm
df_cleaned.head()
```

```python
def suggest_medicines(prefix, names_list):
    prefix = prefix.lower().strip()
    suggestions = [name for name in names_list if name.startswith(prefix)]
    return suggestions
```

```python
# Convert the 'name' column to a cleaned list again (if not already done)
medicine_names = df_cleaned['name'].dropna().str.lower().str.strip().tolist()

# Test the suggestion function
user_input = "a"  # Try with different letters like "az", "al", etc.
suggestions = suggest_medicines(user_input, medicine_names)

# Show suggestions
print("Suggestions:", suggestions)
```
